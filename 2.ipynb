```python
import yfinance as yf

# 提取特斯拉股票数据
tesla_stock = yf.Ticker("TSLA")
tesla_data = tesla_stock.history(period="max")

# 打印前五行数据
print(tesla_data.head())
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd

# 启动 Selenium 浏览器
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://finance.yahoo.com/quote/TSLA/financials")

# 等待页面加载并获取源代码
driver.implicitly_wait(10)
page_source = driver.page_source

# 使用 pandas 解析表格数据
tables = pd.read_html(page_source)

# 假设第一个表格是收入数据
tesla_revenue = tables[0]
print(tesla_revenue.head())

# 关闭浏览器
driver.quit()
```

```python
# 提取游戏驿站股票数据
gamestop_stock = yf.Ticker("GME")
gamestop_data = gamestop_stock.history(period="max")

# 打印前五行数据
print(gamestop_data.head())
```

```python
# 使用 Selenium 获取游戏驿站的收入数据
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://finance.yahoo.com/quote/GME/financials")

# 等待页面加载并获取源代码
driver.implicitly_wait(10)
page_source = driver.page_source

# 使用 pandas 解析表格数据
tables = pd.read_html(page_source)

# 假设第一个表格是收入数据
gamestop_revenue = tables[0]
print(gamestop_revenue.head())

# 关闭浏览器
driver.quit()
```

```python
import matplotlib.pyplot as plt

# 创建图表
fig, ax1 = plt.subplots()

# 绘制特斯拉股票价格
ax1.plot(tesla_data.index, tesla_data['Close'], color='blue')
ax1.set_xlabel('Date')
ax1.set_ylabel('Stock Price', color='blue')

# 使用第二个 y 轴绘制特斯拉收入
ax2 = ax1.twinx()
ax2.plot(tesla_revenue.index, tesla_revenue.iloc[:, -1], color='red')  # 假设最后一列是收入数据
ax2.set_ylabel('Revenue', color='red')

plt.title('Tesla Stock Price and Revenue')
plt.show()
```

```python
# 创建图表
fig, ax1 = plt.subplots()

# 绘制游戏驿站股票价格
ax1.plot(gamestop_data.index, gamestop_data['Close'], color='blue')
ax1.set_xlabel('Date')
ax1.set_ylabel('Stock Price', color='blue')

# 使用第二个 y 轴绘制游戏驿站收入
ax2 = ax1.twinx()
ax2.plot(gamestop_revenue.index, gamestop_revenue.iloc[:, -1], color='red')  # 假设最后一列是收入数据
ax2.set_ylabel('Revenue', color='red')

plt.title('GameStop Stock Price and Revenue')
plt.show()
```
